# Что такое файл
файл - непрерывная последовательность бит (байт) на диске. Нас интересует что файл имеет:
- расположение/адрес/путь на диске
- кодировку
- расширение

О расположении:
любой конкретный символ в книге можно идентефицировать с помощью:
- страницы книги
- строки на странице
- номера символа в строке
при этом какие-то такие символы мы ассоциируем с "главами книги". И когда мы обращаемся к главе "программирование в экономике" чтобы ее прочитать, нам на самом деле надо найти нужный символ на нужной странице.

На диске в компьютере происходит идейно похожий процесс, существует некая файловая система, удобная для человека, в которой у файлов есть имена и иерархия. А операционная система отображает эти человеко-понятные имена в физические адреса на диске.

Можно открыть файлы в гугл колабе и увидеть нашу файловую систему, вы уже должны были сложить датасет в каталог sample_data/

О кодировке и расширении:
изначально, файл на диске это просто последовательность байт, чтобы обозначить тип файла используют "расширение", это все символы после точки. Например у нас мы видим:
- md: markdown, язык разметки текста, кстати все текстовые блоки в этом "ноутбуке" написаны на языке markdown
- json: JavaScript Object Notation, нас не интересует
- csv: Comma-Separated Values, в таких файлах хранят табличные данные
- xlsx: вам знаком -- тоже формат хранения табличных данных, но на стероидах, можно раскрашивать ячейки, вставлять картинки, графики, формулы и тд
Так вот, у файлов которые, например, работают с человекочитаемым текстом есть "кодировка". Кодировка для текстовых файлов это способ превращать последовательность байт в графические символы понятные человеку. Например в кодировке UTF-8.
Например последовательность бит 01011010 это латинская буква Z, а последовательность бит 1101000010010001 это русская буква Б. Так вот файл в человекочитаемой кодировке UTF-8 превращается в текст на вашем экране.

# Функции
Функция в языках программирования близка к математическому смыслу, но не надо думать о них в математическому смысле. Функция в ЯП это способ дать название какому-то фрагменту кода и вызывать "функцию", а не переписывать один и тот же код несколько раз. То есть функции нужны для логической организации кода, то есть для удобства.

В python функция определяется следующей синтаксической конструкцией:

__def__ __ИМЯ_ФУНКЦИИ__ __(__arg1, arg2, ...__)__:<br>
&nbsp;&nbsp;&nbsp;__КОД__<br>
&nbsp;&nbsp;&nbsp;__[return value]__

Также в python можно "аннотировать" тип аргумента функции, например:


```python
def plus(a: int, b: int):
  return a + b
```


````python
def get_dataset_column(column_name: str):
  with open('sample_data/omsk_2025_q4_realestate.csv', newline='\n') as file:
    file.readline()
    column = []
    for row in file:
      for idx, value in enumerate(row[:-1].split('\t')):
        if (idx == column_to_idx[column_name]):
          column.append(value)
  return column

street = get_dataset_column('street')
print(street)
````

В наших данных очевидно есть другие типы данных: целые и дробные числа, даты.

````python
# обозначаем какой тип данных имеет каждая колонка в нашем csv файле
column_types = {
    'street': 'str',
    'year': 'int',
    'floor': 'int',
    'area': 'float',
    'date': 'datetime',
    'amount': 'float'}
````

````python
from datetime import datetime
# функция cast_str_to_type переводит строку к нужному типу данных
def cast_str_to_type(value: str, value_type: str):
  if (value_type == 'str'):
    return value
  if (value_type == 'int'):
    return int(value)
  if (value_type == 'float'):
    return float(value)
  if (value_type == 'datetime'):
    return datetime.strptime(value, "%Y-%m-%d")

  raise ValueError("Value type doesn't supported")
````

````python
def get_dataset_column(column_name: str):
  with open('sample_data/omsk_2025_q4_realestate.csv', newline='\n') as file:
    file.readline()
    column = []

    for row in file:
      for idx, value in enumerate(row[:-1].split('\t')):
        if (idx == column_to_idx[column_name]):
          try:
            #!!!
            value_typed = cast_str_to_type(value, column_types[column_name])

            column.append(value_typed)
          except ValueError:
            break
  return column
````

# Функции от функций

Конечно же аргументом функции может быть другая функция, простой пример из реальности это интеграл (надеюсь всем вам знакомый).

Ниже преведена функция, которая находит приближенное значение интеграла Римана от функции.

````python
def riemann_integral(a: float, b: float, N: int, func):
  if (b <= a):
    raise ValueError("b must be greater then a")
  if (N <= 0):
    raise ValueError("N must be greater then 0")

  delta = (b - a) / N
  value = 0
  for i in range(N):
    #!!!
    value += (func(delta*(i+1)) + func(delta*i)) / 2 * delta

  return value

def func_x(x):
  return x

print('here calc integral for x')
print(riemann_integral(0, 2, 1, func_x))
print(riemann_integral(0, 2, 10, func_x))

def func_x_squared(x):
  return x*x

print('here calc integral for x**2')
print(riemann_integral(0, 2, 1, func_x_squared))
print(riemann_integral(0, 2, 10, func_x_squared))
print(riemann_integral(0, 2, 100, func_x_squared))
print(riemann_integral(0, 2, 1000, func_x_squared))
````

Добавим в нашу функцию, которая получает колонки фильтрацию

````python
def get_dataset_column_with_filter(column_name: str, predicat):
  with open('sample_data/omsk_2025_q4_realestate.csv', newline='\n') as file:
    file.readline()
    column = []
    for row in file:
      for idx, value in enumerate(row[:-1].split('\t')):
        if (idx == column_to_idx[column_name]):
          try:
            value_typed = cast_str_to_type(value, column_types[column_name])
            #!!!
            if (predicat(value_typed)):
              column.append(cast_str_to_type(value, column_types[column_name]))
          except ValueError:
            break
  return column

# lambda x: f(x) это тоже синтаксическая конструкция для создания функций
# функцию, возвращающую True/False будем называть предикатом или фильтром
# можете пользоваться лямбдами, а можете определять функции как делали это ранее
get_dataset_column_with_filter('amount', lambda x: x > 10_000_000)
````
